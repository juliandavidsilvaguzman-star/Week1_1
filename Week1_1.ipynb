## 1. Lógica matemática y función de decisión
Implementación de una regla binaria usando un umbral.

```

import numpy as np

# Función step: convierte un puntaje en una clase (0 o 1)
def step(z):
    if z >= 0:
        return 1
    else:
        return 0

# Pruebas de la función
for z in [1, 0.2, 0, -0.3]:
    print(f'z={z} -> clase={step(z)}')
```

## 2. Perceptrón básico (compuerta AND)
Se implementa un perceptrón con NumPy.

```

# Datos de entrada
X = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=float)

# Pesos y bias
w = np.array([1.0, 1.0])
b = -1.5

# Puntaje y salida
z = X @ w + b
salida = [step(val) for val in z]

print('z:', z)
print('salida:', salida)
```

## 3. Sigmoid vs Step
La sigmoid devuelve probabilidades entre 0 y 1.

```

import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

for z in [-2, -1, 0, 1, 2]:
    print(f'z={z} -> step={step(z)}, sigmoid={sigmoid(z):.3f}')
```

## 4. Mini entrenamiento (ML básico)
Ajuste manual de pesos y bias.

```

y_true = np.array([0,0,0,1])
lr = 0.1

for epoch in range(5):
    print(f'
Epoch {epoch+1}')
    for i in range(len(X)):
        z = X[i] @ w + b
        y_pred = step(z)
        error = y_true[i] - y_pred

        # Actualización
        w += lr * error * X[i]
        b += lr * error
        print(f'x={X[i]}, y_pred={y_pred}, error={error}')
```
